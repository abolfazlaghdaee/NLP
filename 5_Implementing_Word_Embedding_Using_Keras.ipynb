```python
sentences = ['This YouTube channel is perfect',
            'Your videos are quite good',
            'I enjoy this channel',
            'I did not learn much from your videos',
            'Excellent',
            'poor work',
            'it is not a good channel']
```

```python
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.layers import Embedding
from tensorflow.keras.models import Sequential
```

```python
size = 500
```

```python
encoding = [one_hot(sent, size) for sent in sentences]
```

```python
encoding
```

```python
max_len = 8
```

```python
new_encoding  = pad_sequences(encoding, )
new_encoding
```

```python
model = Sequential()
model.add(Embedding(size, 10, input_length = 8, ))
```

```python
model.compile(optimizer = 'adam', loss = 'mse')
```

```python
model.summary()
```

```python
model.predict(new_encoding)
```
